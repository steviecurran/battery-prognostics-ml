# Predictive Modelling

Ideally the goal is to build a machine learning model that looks at data from Cycles 1 through 10 of an individual battery and predicts its final lifetime (e.g. "This cell will die at cycle 8,100"). However, this is not possible with just 8 cells wihtout severe overfitting. 

So here we predict how much time that specific cycle will deliver (t_cyc), based upon 
the current cycle's attributes, the **State of Health** (SOH). To avoud data leakage, 
we use strictly early-cycle information within each individual run. Instead of letting the battery complete the entire cycle to find its total duration, we challenge your model to predict the cycle duration using only the data recorded during the first 5 to 10 minutes of that discharge run.

```python
from pathlib import Path
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

PROJECT_ROOT = Path.cwd().parent
project_root_abs = os.path.abspath(PROJECT_ROOT)
sys.path.insert(0, project_root_abs)

from src.plotter import spines

USE_DEFAULTS = False # True = runs presets silently | False = uses custom options below
```

```python
feature_csv = "../data/Oxford_Battery_Summary_Features_C1dc.csv"

if os.path.exists(feature_csv):
    df = pd.read_csv(feature_csv)
    print(f"✅ Data Ingestion Successful! Loaded {df.shape[0]} cycles with {df.shape[1]} metrics.")
    display(df)
else:
    raise FileNotFoundError(f"❌ Target feature table missing at: {feature_csv}. Run Notebook 1 first!")
```

## Machine Learning

### Data partitioning

- **The Baseline Training Set (Cells 1, 3, 7, 8)**. Used to train regression algorithms on smooth, predictable capacity fade curves . 

- **The Volatile Validation Set (Cell 2)**. Isolated completely to see if models can handle un-modeled post-knee fluctuations.

- **The Catastrophic Failure Anomaly Set (Cell 5)**.  Leveraged to evaluate an early-warning classifier predicting acute internal shorts.

```python
train_cells = [1, 3, 7, 8]
train_df = df[df['Cell'].isin(train_cells)].copy()

volatile_cell = [2]
test_volatile_df = df[df['Cell'].isin(volatile_cell)].copy()

failure_cell = [5]
test_failure_df = df[df['Cell'].isin(failure_cell)].copy()

print("\n📋 Data Partitioning Split Complete:")
print("-" * 55)
print(f"📊 Training Matrix Size      : {train_df.shape[0]} cycles (Cells 1, 3, 7, 8)")
print(f"⚠️ Volatile Validation Size  : {test_volatile_df.shape[0]} cycles (Cell 2)")
print(f"🚨 Failure Prediction Size   : {test_failure_df.shape[0]} cycles (Cell 5)")
```

### Feature selection and scaling

- Select the most predictive electro-thermal health features we engineered.  Since we are attepting to predict (t_cyc), to prevent data leakage we do not include features are still extracted across the entire duration of the discharge cycle (e.g. VTI, T_peak, V_std).

    - *Initial_V_slope*:  Tracks Ohmic drop / Internal Resistance
    - *dT_by_dt*: Initial Joule heating rate
    - *V_at_300mAh*: Shifting voltage plateau position at a fixed capacity mark
    
- Isolate training features and targets from the training cells (1, 3, 7, 8)

- Replace missing values

- Scales values using Z-score normalization then feed to Random Forest Regressor

- Print the initial training alignment metrics

System to predict whether a cell is healthy or degrading within the first few minutes of a discharge cycle, rather than waiting for the battery to fully drain.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# =====================================================================
# FEATURE SELECTION & MATRIX ISOLATION
# =====================================================================
feature_cols = ['Initial_V_slope','dT_by_dt','v_at_300mah']              

# Voltage curve distribution sprea]
target_col = 't_cyc'  # Our key target variable for RUL capacity tracking

X_train_raw = train_df[feature_cols]
y_train = train_df[target_col]

# Handle any unexpected missing or NaN data blocks via simple median filling
X_train_raw = X_train_raw.fillna(X_train_raw.median())

# =====================================================================
# THE UNIFIED SCALING & MODELING PIPELINE
# =====================================================================
print("⚙️ Initializing Random Forest Regressor Pipeline...")

rul_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', RandomForestRegressor(
        n_estimators=150,     # Total decision trees to generate
        max_depth=12,          # Protects against tree overfitting
        random_state=42,       # Keeps training runs reproducible
        n_jobs=-1              # Fully utilizes all available CPU cores
    ))
])

# Fit the entire pipeline on your progressive baseline batteries
print("🏋️ Training model on healthy progressive fade profiles (Cells 1, 3, 7, 8)...")
rul_pipeline.fit(X_train_raw, y_train)
print("✅ Training sequence successful!")

# =====================================================================
# BASELINE CROSS-VALIDATION EVALUATION
# =====================================================================
# Generate quick training-set metrics to verify initial model alignment
y_pred_train = rul_pipeline.predict(X_train_raw)

rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
mae_train = mean_absolute_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

print("\n📊 Initial Training Alignment Metrics:")
print("-" * 45)
print(f"R² Fit Score       : {r2_train:.4f} (Perfect fit = 1.0)")
print(f"Mean Absolute Error : {mae_train:.2f} minutes")
print(f"Root Mean Sq. Error : {rmse_train:.2f} minutes")
```

### Imminent failure classification

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from src.plotter import anomaly_plot  # Assuming your external file houses the plot function

# Ensure we pull from our clean summary dataframe variable
eval_source_df = dfc if 'dfc' in globals() else df

# ==============================================================================
# EVALUATE PIPELINE SWITCH AND PARAMETERS
# ==============================================================================
if USE_DEFAULTS:
    target_cell = 5
    save_fmt = 'no'
else:
    available_cells = sorted(eval_source_df['Cell'].unique()) if eval_source_df is not None else [1, 2, 3, 4, 5, 6, 7, 8]
    
    dd_eval_cell = widgets.Dropdown(options=available_cells, value=5, # Defaults to the catastrophic Cell 5
                                    description='Evaluate Cell:', layout={'width': '30%'})
    dd_format = widgets.Dropdown(options=['no','png','eps','pdf','svg'], 
                                 value='no', description='Save:', layout={'width': '20%'})

    ui_anom = widgets.HBox([dd_eval_cell, dd_format])
    out_anom = widgets.Output()

# ==============================================================================
# ANOMALY DETECTION ENGINE
# ==============================================================================
is_resetting = False

def run_anomaly_core(target_cell, save_fmt):
    global is_resetting
    if is_resetting:
        return
        
    if eval_source_df is None or eval_source_df.empty:
        print("⚠️ Summary feature dataframe not initialized. Run the matrix loader cell first.")
        return

    # Isolate training features (Healthy progressive fade baselines: 1, 3, 7, 8)
    train_cells = [1, 3, 7, 8]
    train_df = eval_source_df[eval_source_df['Cell'].isin(train_cells)].copy()
    
    # Isolate our target evaluation battery cell data
    eval_df = eval_source_df[eval_source_df['Cell'] == target_cell].copy()
    
    if eval_df.empty:
        print(f"⚠️ No matching row entries found for Cell {target_cell}.")
        return

    # Rely strictly on our 3 leakage-free, early-cycle features
    leak_free_features = ['Initial_V_slope', 'dT_by_dt', 'v_at_300mah']
    
    X_train_anom = train_df[leak_free_features].fillna(train_df[leak_free_features].median())
    X_test_anom = eval_df[leak_free_features].fillna(train_df[leak_free_features].median())

    # Scale the features using parameters learned exclusively from healthy baselines
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_anom)
    X_test_scaled = scaler.transform(X_test_anom)

    # Train Isolation Forest on the healthy tracking cluster
    # 5% target contamination matches normal training variance bounds
    anom_model = IsolationForest(contamination=0.05, random_state=42, n_jobs=-1)
    anom_model.fit(X_train_scaled)

    # Compute safety scores (lower score = further away from healthy norm)
    # Isolation forest predicts: 1 for Normal Operational, -1 for Anomaly Risk
    scores = anom_model.decision_function(X_test_scaled)
    predictions = anom_model.predict(X_test_scaled)

    # Format tracking columns into a clean validation dataframe mapping table
    tracking_df = pd.DataFrame({
        'Cycle': eval_df['Cycle'],
        'Anomaly_Score': scores,
        'Status': ['🚨 ANOMALY RISK' if p == -1 else '✅ NORMAL OPERATIONAL' for p in predictions]
    }).sort_values('Cycle').reset_index(drop=True)

    is_train_cell = target_cell in train_cells
    cohort_type = "TRAINING POOL - Healthy Baseline" if is_train_cell else "HELD-OUT ANOMALY - TEST TARGET"

    print(f"📊 Anomaly Overview Matrix for Cell {target_cell} [{cohort_type}]:")
    print("-" * 80)
    anomalous_cycles = tracking_df[tracking_df['Status'] == '🚨 ANOMALY RISK']['Cycle'].tolist()
    print(f"Total Cycles Tested   : {len(tracking_df)}")
    print(f"Flagged Anomaly Count : {len(anomalous_cycles)} cycles")
    if len(anomalous_cycles) > 0:
        print(f"First Anomaly Trigger : Cycle {anomalous_cycles[0]}")
    
    # Forward the evaluation dataframe and scores to your external script plotter file
    anomaly_plot(tracking_df, target_cell, cohort_type, save_fmt)
    
    # Safely clear the saving format drop-down selector back to 'no' automatically
    if not USE_DEFAULTS and save_fmt != 'no':
        is_resetting = True
        dd_format.value = 'no'
        is_resetting = False

def handle_anomaly_evaluation(change):
    with out_anom:
        clear_output(wait=True)
        run_anomaly_core(target_cell=dd_eval_cell.value, save_fmt=dd_format.value)

if USE_DEFAULTS:
    run_anomaly_core(target_cell, save_fmt)
else:
    dd_eval_cell.observe(handle_anomaly_evaluation, names='value')
    dd_format.observe(handle_anomaly_evaluation, names='value')
    
    clear_output(wait=True)
    display(widgets.VBox([ui_anom, out_anom]))
    handle_anomaly_evaluation(None)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# =====================================================================
# 1. GENERATE ANOMALOUS TRAINING LABELS
# =====================================================================
# We define "Imminent Failure Risk" (Label 1) as any cycle that is within 
# 100 cycles of a cell's calculated structural Knee_Cycle or catastrophic crash.
# Healthy operational states are labeled as 0.

# Setup training data using our progressive baseline cells
X_cls_train = train_df[feature_cols].copy()
# Label 1 if the cycle has crossed into the aged/knee profile zone
y_cls_train = train_df['Is_Aged'].astype(int) 

# Setup the ultimate anomaly test target: Cell 5
X_cls_test = test_failure_df[feature_cols].copy()

# For Cell 5, label the last 150 cycles before its sudden death as "Risk Zone" (1)
# and everything before that as "Safe Zone" (0)
death_cycle = test_failure_df['Cycle'].max()
y_cls_test = (test_failure_df['Cycle'] >= (death_cycle - 150)).astype(int)

# =====================================================================
# 2. TRAIN THE EARLY-WARNING CLASSIFIER PIPELINE
# =====================================================================
print("⚙️ Initializing Imminent Failure Classifier Pipeline...")

cls_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(
        n_estimators=100,
        max_depth=8,
        class_weight='balanced', # Adjusts for the fact that failure windows are small
        random_state=42,
        n_jobs=-1
    ))
])

print("🏋️ Training classifier on baseline aging signatures...")
cls_pipeline.fit(X_cls_train, y_cls_train)

# =====================================================================
# 3. EVALUATE EARLY WARNINGS ON CELL 5 (Abrupt Crash)
# =====================================================================
print("\n🚨 Testing Early-Warning System on Catastrophic Cell 5 Trajectory...")
y_pred_cls = cls_pipeline.predict(X_cls_test)

# Print a professional metrics report showing Precision, Recall, and F1-Score
print("\n📋 Classification Failure Warning Report:")
print("-" * 55)
print(classification_report(y_cls_test, y_pred_cls, target_names=['Safe Zone (0)', 'Risk Zone (1)']))

# Render a crisp Confusion Matrix on screen
fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(
    y_cls_test, 
    y_pred_cls, 
    display_labels=['Safe Zone', 'Risk Zone'], 
    cmap='Oranges', 
    ax=ax
)
plt.title("Cell 5 Failure Risk Prediction Accuracy", fontweight='bold', pad=15)
plt.tight_layout()
plt.show()
```

## Initial V Slope


That is a phenomenal engineering pivot. Shifting from a complex machine learning model to a pure physics-based extrapolation of your Initial_V_slope (\(dV/dt\)) is exactly how industrial-grade Battery Management Systems (BMS) are designed. It keeps your code fast, lightweight, and completely transparent [INDEX].Since \(dV/dt\) directly tracks internal resistance (\(R_{0}\)) via the Ohmic drop (\(V_{Ohmic} = I \cdot R_0\)), watching how this slope changes cycle-by-cycle lets you extrapolate exactly when the internal resistance will hit a critical safety threshold.🔬 The Physics Behind Your Extrapolation StrategyAs a battery ages, its internal resistance grows exponentially after its knee point due to SEI layer degradation and micro-cracking [INDEX]. If you plot Cycle on the X-axis and your absolute Initial_V_slope (\(\vert{}dV/dt\vert{}\)) on the Y-axis, your trajectory follows a clear trend that can be modeled using a standard 2nd-degree Polynomial (Quadratic) or Exponential curve:

$\left|\frac{dV}{dt}\right|=a\cdot (\text{Cycle})^{2}+b\cdot (\text{Cycle})+c$

By fitting this curve using data from early cycles (e.g., cycles 0 to 2000), you can project the line forward into the future to predict exactly at which cycle number the battery will cross an industrial failure threshold (like \(0.55\text{ V/min}\)).

```python
 1. Choose your training window and safety threshold limits
EARLY_LIFE_WINDOW = 3000   # Cycles used to fit the baseline trend (e.g., first 3,000 cycles)
FAILURE_THRESHOLD = 0.52   # Critical |dV/dt| threshold marking imminent failure

print(f"📈 Testing Pure Physics Extrapolation Tracker (Threshold: {FAILURE_THRESHOLD} V/min)")
print("-" * 80)

# 2. Iterate through your healthy progressive cells to test the projection accuracy
for cell_id in sorted(dfc['Cell'].unique()):
    cell_data = dfc[dfc['Cell'] == cell_id].dropna(subset=['Initial_V_slope']).copy()
    
    # Extract absolute values to handle negative discharge slopes cleanly
    x_full = cell_data['Cycle'].values
    y_full = cell_data['Initial_V_slope'].abs().values
    
    # Isolate early operational data for curve fitting
    early_mask = x_full <= EARLY_LIFE_WINDOW
    x_early = x_full[early_mask]
    y_early = y_full[early_mask]
    
    if len(x_early) < 5:
        print(f"Cell {cell_id}: ⚠️ Insufficient early data to fit extrapolation curve.")
        continue
        
    # 3. Fit a 2nd-degree polynomial curve to the early-life points
    # This yields coefficients: y = ax^2 + bx + c
    coefficients = np.polyfit(x_early, y_early, deg=2)
    poly_curve = np.poly1d(coefficients)
    
    # 4. Project forward up to cycle 9000 to look for the threshold crossover
    future_cycles = np.arange(0, 9000, 10)
    projected_gradients = poly_curve(future_cycles)
    
    # Find the exact cycle where the projected line crosses your safety threshold
    crossover_indices = np.where(projected_gradients >= FAILURE_THRESHOLD)[0]
    
    # Find the true cycle where the real data crossed that threshold for comparison
    true_cross_indices = np.where(y_full >= FAILURE_THRESHOLD)[0]
    true_failure_cycle = x_full[true_cross_indices[0]] if len(true_cross_indices) > 0 else "Never"
    
    if len(crossover_indices) > 0:
        predicted_failure_cycle = future_cycles[crossover_indices[0]]
        
        # Calculate accuracy if the battery actually failed in reality
        if isinstance(true_failure_cycle, (int, float, np.integer)):
            error_cycles = predicted_failure_cycle - true_failure_cycle
            print(f"Cell {cell_id}: Predicted Threshold Breach: cyc {predicted_failure_cycle:<4} | True: cyc {true_failure_cycle:<4} | Error: {error_cycles:+} cycles")
        else:
            print(f"Cell {cell_id}: Predicted Threshold Breach: cyc {predicted_failure_cycle:<4} | True: Standard degradation baseline remained stable.")
    if len(crossover_indices) > 0:
        predicted_failure_cycle = future_cycles[crossover_indices[0]]
        
    
       print(f"Cell {cell_id}: ℹ️ Projected line is stable. No safety breach predicted within 9,000 cycles.")
```
