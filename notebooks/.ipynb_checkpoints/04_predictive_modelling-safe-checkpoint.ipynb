# Predictive Modelling

Ideally the goal is to build a machine learning model that looks at data from Cycles 1 through 10 of an individual battery and predicts its final lifetime (e.g. "This cell will die at cycle 8,100"). However, this is not possible with just 8 cells wihtout severe overfitting. 

So here we predict how much time that specific cycle will deliver (t_cyc), based upon 
the current cycle's attributes, the **State of Health** (SOH). To avoud data leakage, 
we use strictly early-cycle information within each individual run. Instead of letting the battery complete the entire cycle to find its total duration, we challenge your model to predict the cycle duration using only the data recorded during the first 5 to 10 minutes of that discharge run.

```python
default_csv = "../data/Oxford_Battery_Dataset_Test=C1dc.csv"
if os.path.exists(default_csv):
    print(f"📦 Reading pre-compiled file from disk:\n {default_csv}")
    df = pd.read_csv(default_csv)
    print(f"✅ Success! Loaded {len(df):,} rows into memory.")
    display(df)
```

```python
from scipy.stats import linregress
import ipywidgets as widgets
from IPython.display import display, clear_output
from scipy.integrate import trapz
pd.set_option('display.max_columns', None)

cells = [i for i in range(1,df['Cell'].iloc[-1] + 1)]   

def build_early_features_ui(df, window,prev):
    """Processes battery parameters reactively based on the active UI window choice."""
    if df is None or df.empty:
        print("⚠️ Master dataframe array context is currently uninitialized.")
        return
        
    print(f"⚙️ Running Active Feature Extraction... Target Window: {window} minutes.")
    print("⏳ Vectorizing cycle groups (please hold)...")
    
    rows = []
    # Loop over your cell and cycle combinations smoothly
    for (cell, cycle), g in df.groupby(["Cell", "Cycle"]):
        g = g.sort_values("Time_min")
        early = g[g["Time_min"] <= window]
        
        # Guard clause ensures there are enough physical data points to run linregress
        if len(early) < 20:
            continue

        vfit = linregress(early["Time_min"], early["Voltage_V"])
        tfit = linregress(early["Time_min"], early["Temperature_C"])
        coeff = np.polyfit(early.Time_min,early.Voltage_V,2)
        auc = trapz(early.Voltage_V,early.Time_min
)
        
        rows.append({"Cell": cell,"Cycle": cycle,
                     "V_start": early["Voltage_V"].iloc[0],
                     "V_end": early["Voltage_V"].iloc[-1],
                     "V_mean": early["Voltage_V"].mean(), 
                      "V_std": early["Voltage_V"].std(),
                      "V_slope": vfit.slope, 
                      "V_curve":coeff[0], # Quadratic term for curvature
                      "VTI": auc,
                      "T_start": early["Temperature_C"].iloc[0],
                      "T_end": early["Temperature_C"].iloc[-1], 
                      "T_mean": early["Temperature_C"].mean(), 
                      "T_std": early["Temperature_C"].std(),
                       "T_slope": tfit.slope,
                       "q_end": abs(early["Charge_mAh"].iloc[-1]),
                       "t_end": early["Time_min"].iloc[-1],
                       "t_cyc": g["Time_min"].max() # ADD THE TARGET FOR THAT CYCLE
        })

    # Convert results into a globally accessible feature matrix
    global early_features_df
    early_features_df = pd.DataFrame(rows)
    early_features_df["FinalCycle"] = ( early_features_df.groupby("Cell")["Cycle"].transform("max"))
    early_features_df["RUL"] = (early_features_df["FinalCycle"]- early_features_df["Cycle"])
    
    print(f"✅ Complete! Extracted matrix layout: {early_features_df.shape} rows and columns.")
    print("Showing new dataframe for Cell %d below" %(prev))
    #display(early_features_df.head(5))
    tmp = early_features_df[early_features_df['Cell']==prev]
    display(tmp)

# ==============================================================================
# INTERACTIVE USER INTERFACE MECHANICS
# ==============================================================================
dd_win = widgets.Dropdown(options=[1, 2, 5, 10, 20, 60], value=5,
    description='Time window in minutes:',style={'description_width': 'initial'},
    layout={'width': '35%'})

dd_prev = widgets.Dropdown(options=cells, value=2,
    description='Select a cell to preview below:',
    style={'description_width': 'initial'},
    layout={'width': '35%'})

ui = widgets.HBox([dd_win,dd_prev])

# ==============================================================================
# ROUTING AND EXECUTION MANAGEMENT (Respects USE_DEFAULTS)
# ==============================================================================
early_features_df = pd.DataFrame() # Global container initial placeholder

if USE_DEFAULTS:
    print("⚙️ Running Mode: [DEFAULTS ACTIVE] | Target Window: 5 minutes")
    # Instant, static bypass execution path
    build_early_features_ui(df, window=5, prev=2)
else:
    # Clear the layout space cleanly
    clear_output(wait=True)
    
    # Crucial: Use widgets.fixed(df) to feed the data directly to the background thread
    out = widgets.interactive_output(
        build_early_features_ui, {
            'df': widgets.fixed(df),'window': dd_win, 'prev':dd_prev
        }
    )
    
    display(ui, out)
```

### Considerations

"If I manufacture a new battery tomorrow, can this model estimate its health from the first five minutes of discharge?"

When applying machine learning to time series data, we have to be wary the standard train/test random split causing leakage. With  we may still be feeding the model the history of the cell, from which the algorithm can simply guess the prediction. While this will give excellent results, it will not provdide a reliable prediction for a battery the model has never seen before.

You are entirely correct, and that is a profound machine learning insight.If you use a standard randomized row split (train_test_split), data points from the same battery cell end up in both the training set and the test set. Because sequential cycles are highly correlated, a model like a Random Forest doesn't learn how a battery degrades; it just interpolates based on neighboring cycles it has already memorized. This is a severe form of data leakage.

So here we split the data by cell and attempt to predict the behavior of an unseen cell.


s that "leakage"?

Technically, no.

You're not leaking future labels or future features.

I'd call it dependency leakage or trajectory leakage.

The training and test examples are not independent.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings

warnings.simplefilter(action='ignore', category=np.RankWarning)

# Global memory caches to store arrays loop-free
cached_summary_results = pd.DataFrame()
cached_predictions_df = pd.DataFrame()

def train_and_cache_predictions(dfc):
    """Executes LOGO cross-validation across all models and targets exactly once

    to cache features for instant UI plotting.
    """
    if dfc is None or dfc.empty:
        print("⚠️ Summary early feature matrix array is currently uninitialized.")
        return
        
    win = round(dfc["t_end"].iloc[0], 0)
    df = dfc.copy()

    regressors = {
        "Dummy": DummyRegressor(),
        "Linear": LinearRegression(),
        "Ridge": Ridge(),
        "Random Forest": RandomForestRegressor(random_state=42, n_jobs=-1),
        "Gradient Boosting": HistGradientBoostingRegressor(random_state=42)
    }

    groups = df["Cell"]
    cycles = df["Cycle"]
    
    # We loop over BOTH targets sequentially right here to cache everything at once
    targets_to_process = ['t_cyc', 'RUL'] if 'RUL' in df.columns else ['t_cyc']
    
    all_metrics = []
    all_predictions = []

    print(f"🏋️ Initiating Global Leave-One-Cell-Out Cross-Validation (Window: {win:.0f} min)...")
    print("⏳ Fitting regressors across all cell matrices (this runs ONCE)...")

    for target in targets_to_process:
        X = df.drop(columns=['t_cyc', "Cell", "Cycle", "FinalCycle", "RUL", "t_end"], errors='ignore')
        y = df[target]
        logo = LeaveOneGroupOut()

        for train_idx, test_idx in logo.split(X, y, groups):
            test_cell = groups.iloc[test_idx].iloc[0]
            test_cycles = cycles.iloc[test_idx]

            X_train = X.iloc[train_idx].copy()
            X_test = X.iloc[test_idx].copy()
            y_train = y.iloc[train_idx]
            y_test = y.iloc[test_idx]

            scaler = StandardScaler()
            X_train = scaler.fit_transform(X_train)
            X_test = scaler.transform(X_test)

            for key, model in regressors.items():
                model.fit(X_train, y_train)
                y_pred = model.predict(X_test)
                
                rms = np.sqrt(mean_squared_error(y_test, y_pred))
                mae = mean_absolute_error(y_test, y_pred)
                r2 = r2_score(y_test, y_pred)
                r_val = np.corrcoef(y_test, y_pred)[0, 1] if np.std(y_pred) > 0 and np.std(y_test) > 0 else 0.0
                
                p = np.polyfit(y_pred, y_test, 1)
                
                # We divide by the mean of the actual test cell values to find the fraction, 
                # then multiply by 100 to convert it into a readable percentage.
                baseline_val = np.mean(y_test)
                rms_frac = (rms / baseline_val)
                mae_frac = (mae / baseline_val) 

                all_metrics.append({
                    "Target": target, "Model": key, "Cell": test_cell, 
                    "RMSE": rms, "MAE": mae, 
                     "RMSE_frac": rms_frac, "MAE_frac": mae_frac, 
                    "Pearson_coeff": r_val, 
                    "R2": r2, "m": p[0], "c": p[1]
                })

                all_predictions.append(pd.DataFrame({
                    "Target": target, "Model": key, "Cell": test_cell, 
                    "Cycle": test_cycles.values, "Actual": y_test.values, 
                    "Predicted": y_pred, "t_end": win  
                    
                }))

    global cached_summary_results, cached_predictions_df
    cached_summary_results = pd.DataFrame(all_metrics)
    cached_predictions_df = pd.concat(all_predictions, ignore_index=True)
    print("✅ Complete! All model metrics and predictions successfully cached in memory.")

# Run the training operations instantly
train_and_cache_predictions(early_features_df)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from src.plotter import ML_plot

def run_instant_dashboard(target, chosen_model):
    if cached_predictions_df.empty:
        print("⚠️ Prediction cache is empty. Please run the training cell above first.")
        return
        

    tgt_metrics = cached_summary_results[cached_summary_results['Target'] == target]
    summary = (
        tgt_metrics.groupby("Model").agg(
            RMSE_mean=("RMSE", "mean"), RMSE_std=("RMSE", "std"),
            MAE_mean=("MAE", "mean"), MAE_std=("MAE", "std"),
            R2_mean=("R2", "mean"), R2_std=("R2", "std"),
            gradient_mean=("m", "mean"), gradient_std=("m", "std"),
            intercept_mean=("c", "mean"), intercept_std=("c", "std")
        ).sort_values("RMSE_mean")
    )
    
    print(f"\n📊 Summary LOCO Cross-Validation Metrics (Target View: {target})")
    print("-" * 90)
    display(summary)
    
    plot_slice = cached_predictions_df[cached_predictions_df['Target'] == target]
    
    ML_plot(plot_slice, target, chosen_model)


if USE_DEFAULTS:
    run_instant_dashboard(target="t_cyc", chosen_model="Random Forest")
else:
    # Build clean dropdown selector layouts side-by-side
    dd_targ = widgets.Dropdown(options=['t_cyc', 'RUL'], value='t_cyc',
                               description='Target:', layout={'width': '25%'})
    
    dd_model = widgets.Dropdown(options=['Dummy', 'Linear', 'Ridge', 'Random Forest', 'Gradient Boosting'], 
                                value='Random Forest', description='Model View:', layout={'width': '30%'})
    
    menu = widgets.HBox([dd_targ, dd_model])
    ui = widgets.VBox([menu])
    
    clear_output(wait=True)
    
    # Interactive output maps straight to our lightweight calculator function
    out = widgets.interactive_output(
        run_instant_dashboard,
        {
            'target': dd_targ,
            'chosen_model': dd_model
        }
    )
    display(ui, out)
```

EXPLAIN METRICS

fractional avlues

To express the RMSE and MAE as fractional (or percentage) errors, we need to divide the error values by a baseline reference value. In battery health tracking, the most robust engineering baseline is the initial pristine value of the cell's lifespan—meaning the first recorded cycle time (\(t_{\text{cyc}}\) at Cycle 0) or the maximum possible Remaining Useful Life (RUL).For your dataset, the initial value sits at approximately 50 to 60 minutes for cycle time (\(t_{\text{cyc}}\)), and up to 8,500 cycles for RUL. Converting these errors to fractional percentages gives reviewers an instant, normalized sense of accuracy (e.g., "The model predicts battery life within a 1% error margin").

## Meta analysis

This kind of meta-analysis over multiple random shuffles is exactly how you prove that a machine learning model's performance is stable, reproducible, and not just a fluke from a lucky random data split.

Zero History Leakage: Because it strictly uses LeaveOneGroupOut, the model is always tested on a completely hidden battery cell whose history is never shown during training. The algorithm cannot cheat by memorizing adjacent cycles.True Meta-Analysis Metrics: By calculating RMSE_std and R2_std over multiple cross-validation loops, your report can definitively prove if your model's predictive power is stable regardless of random data omissions.
