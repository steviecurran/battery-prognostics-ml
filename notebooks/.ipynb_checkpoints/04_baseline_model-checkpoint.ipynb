# Baseline Model

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

feature_csv = "../data/Oxford_Battery_Summary_Features_C1dc.csv"

if os.path.exists(feature_csv):
    ml_df = pd.read_csv(feature_csv)
    print(f"✅ Data Ingestion Successful! Loaded {ml_df.shape[0]} cycles with {ml_df.shape[1]} metrics.")
else:
    raise FileNotFoundError(f"❌ Target feature table missing at: {feature_csv}. Run Notebook 1 first!")

# Separate metadata fields from purely numeric mathematical predictors
# We drop identifier labels and target variables that shouldn't leak into correlation checks
metadata_cols = ['Cell', 'Cycle', 'Test_Type', 'Total_Life_Cycles', 'Pct_Lifespan_Reached', 'Is_Aged']
numeric_features = ml_df.drop(columns=metadata_cols, errors='ignore')

correlation_matrix = numeric_features.corr()
plt.figure(figsize=(14, 12))
font = 14

cmap = sns.diverging_palette(230, 20, as_cmap=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(correlation_matrix,mask=mask, cmap=cmap, vmax=1.0, vmin=-1.0, center=0,
    square=True, linewidths=.7, 
    cbar_kws={"shrink": .75, "label": "Pearson Correlation Coefficient ($r$)"},
    annot=True, fmt=".2f", annot_kws={"size": 8, "weight": "bold"})

plt.title("Engineered Feature Correlation Matrix (Predictive Modeling Input Screen)", fontsize=font, pad=20)
plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
plt.tight_layout()

save_format = 'png' 
if save_format != 'no':
    out_dir = "../images/"
    os.makedirs(out_dir, exist_ok=True)
    plt.savefig(os.path.join(out_dir, f"feature_correlation_heatmap.{save_format}"), dpi=300, bbox_inches='tight')
    print(f"💾 Feature correlation heatmap exported successfully to {out_dir}")

plt.show()
```
